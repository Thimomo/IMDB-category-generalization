```python
import pandas as pd
```

```python
X_train = pd.read_csv('IMDb_train.csv')
X_test = pd.read_csv('IMDb_test.csv')
y_train = pd.read_csv('y_train.csv')
y_test = pd.read_csv('y_test.csv')
train_sets = [X_train, X_test]
```

```python
columns = ['url', 'collections', 'label']
for i in range(len(train_sets)):
    existing_columns = set(train_sets[i].columns).intersection(columns)
    if existing_columns:
        train_sets[i] = train_sets[i].drop(axis=0, columns=existing_columns)
```

```python
def rating(text):
    if text == 'TV-PG':
        return 0
    else:
        return 1

for train_set in train_sets:   
    train_set['content rating'] = train_set['content rating'].apply(rating)
```

```python
def count_minutes(text):
    count = 0
    text = str(text)
    for i in range(len(text)):
        if i < 2:
            continue
        if text[i] == 'H' and text[i-2].isdigit():
            count += int(text[i-2:i]) * 60
        elif text[i] == 'H' and text[i-1].isdigit():
            count += int(text[i-1:i]) * 60
        elif text[i] == 'M' and text[i-2].isdigit():
            count += int(text[i-2:i])
        elif text[i] == 'M' and text[i-1].isdigit():
            count += int(text[i-1:i])
    return count
            

for train_set in train_sets:
    train_set['duration'] = train_set['duration'].apply(count_minutes)
```

```python
max_rating = train_sets[0]['IMDb rating'].max()

def normalize_score(score):
    return score / max_rating
```

```python
for train_set in train_sets:
    train_set['IMDb rating'] = train_set['IMDb rating'].apply(normalize_score)
```

```python
bool_columns = ['genre', 'director', 'stars', 'writer']

for column in bool_columns:
    train_sets[0][column] = train_sets[0][column].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
    train_sets[0][column] = train_sets[0][column].apply(lambda x: [item.strip() for item in x.split(',')])

    new_columns = {}

    for i, row in train_sets[0].iterrows():
        for item in row[column]:
            new_column = item.strip()
            if new_column not in new_columns:
                new_columns[new_column] = [0] * len(train_sets[0])
            new_columns[new_column][i] = 1

    train_sets[0] = pd.concat([train_sets[0], pd.DataFrame(new_columns)], axis=1)

    train_sets[0] = train_sets[0].drop(columns=[column], axis=1)
```
