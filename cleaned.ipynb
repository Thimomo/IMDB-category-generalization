```python
import pandas as pd
import numpy as np
import re


file_path = 'Datasets/cleaned_X_train.csv'
df = pd.read_csv(file_path)

def process_list_column(column):
    def clean_and_eval(x):
        try:
           
            x = re.sub(r'\s+', ',', x.strip())  
            x = f"[{x}]"  
            elements = eval(x)  
            return np.mean(elements)  
        except Exception as e:
            print(f"Fout bij het verwerken van de waarde: {x}. Foutmelding: {e}")
            return np.nan

    return column.apply(clean_and_eval)

df['title'] = process_list_column(df['title'])
df['IMDb description'] = process_list_column(df['IMDb description'])

output_file_path = 'Datasets/cleaned_X_train2.csv'
df.to_csv(output_file_path, index=False)

print(f"De gegevens zijn succesvol opgeslagen in: {output_file_path}")
```
