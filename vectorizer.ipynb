```python
import glob
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train = pd.read_csv('IMDb_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('IMDb_test.csv')
y_test = pd.read_csv('y_test.csv')
target_names = y_test.columns.values
```

```python
def file_list(X):

    base_directory = 'YOUR/PATH/HERE/*'
    l = []

    for label in X['label']:
        pattern = os.path.join(base_directory, '*', label)
        for file_path in glob.iglob(pattern, recursive=True):
            l.append(file_path)
    return l
```

```python
def test_fit(X, vectorizer):
    files = file_list(X)
    vector = vectorizer.transform((open(filename).read() for filename in files))
    return vector
```
