```python
import glob
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train = pd.read_csv('Datasets/IMDb_train.csv') # Input features
y_train = pd.read_csv('Datasets/y_train.csv') # Target 

X_test = pd.read_csv('Datasets/IMDb_test.csv')
y_test = pd.read_csv('Datasets/y_test.csv')

target_names = y_test.columns.values
```

```python
def file_list(X):

    base_directory = 'Datasets/*'
    l = []

    for description in X['IMDb description']:
        pattern = os.path.join(base_directory, '*', description)
        for file_path in glob.iglob(pattern, recursive=True):
            l.append(file_path)
            
    print(l)
    return l
```

```python
def test_fit(X, vectorizer):
    files = file_list(X)
    vectorizer.fit(open(filename).read() for filename in files)
    vector = vectorizer.transform((open(filename).read() for filename in files))
    return vector
```
