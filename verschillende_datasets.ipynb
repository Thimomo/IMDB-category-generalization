```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
contraction_mapping = {
    "wasn't": "was not",
    "didn't": "did not",
    "he's":"he is",
    "doesn't": "does not",
    "hadn't": "had not",
    "hasn't": "has not",
    "won't": "will not",
    "don't": "Do not",
    "it's": "it is",
    "isn't": "is not",
    "can't": "cannot",
    "that's": "that is",
    "i'm": "I am",
    "i'll": "I will",
    "it'll": "it will",
    "couldn't": "could not",
    "they'll": "they will",
    "you'll": "you will",
    "wouldn't": "would not",
    "i've": "I have",
    "that's": "that is",
    "Ma'am": "madam",
    "you're": "you are",
    "she'll": "she will",
    "there's": "there is",
    "don't": "do not",
    "don't": "do not",
    "shan't": "shall not",
    "one's": "ones",
    "she's": "she is",
    "let's": "let us",
    "you'd": "you would",
    "meanwhile, back o...":"",
    "&apos;":"'",
    "apos":"",
    "capt.":"captain",
    "capt" : "captain",
    "&quot;good vs.":"",
    "valentine's Day":"Valentine's_Day",
    "&quot;":"'",
    " ar...":"",
    "throug...":"through.",
    "unite...":"United States.",
    " however, the Robot Dev...":"", 
    "tota...":"totally.",
    "dime...":"dimension.",
    "transporter cy...":"transporter.",
    "reme...":"remembers.",
    "begi...":"begins.",
}

def expand_contractions_in_series(series, contraction_mapping):
    """
    Expand contractions in a Pandas Series.
    """
    def expand_contractions(text):
        for word in word_tokenize(text):
            if word.lower() in contraction_mapping.keys(): 
                text = text.replace(word, contraction_mapping[word.lower()])                    
        return text                 
    
    return series.apply(expand_contractions)

def clean_description(description):
    """
    Tokenizes, lowercases, removes non-alphanumeric tokens, and excludes stopwords.
    """
    if not isinstance(description, str):  
        description = ""
    return ' '.join([word.lower() for word in word_tokenize(description) if word.isalnum() and word.lower() not in stop_words])
```

```python
text_columns = ['title', 'IMDb description']

for column in text_columns:
    for i in range(len(train_sets)):
        train_sets[i][column] = expand_contractions_in_series(train_sets[i][column], contraction_mapping)
        train_sets[i][column] = train_sets[i][column].apply(clean_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
max_description = train_sets[0]['IMDb description'].apply(lambda x: len(str(x).split())).max()
print(max_description)
max_title = train_sets[0]['title'].apply(lambda x: len(str(x).split())).max()
print(max_title)
```

```python
tfidf = TfidfVectorizer(max_features=100)

text_columns = ['title', 'IMDb description']

for column in text_columns:
    for i in range(len(train_sets)):
        results = tfidf.fit_transform(train_sets[i][column]).toarray()
        tfidf_df = pd.DataFrame(results, columns=[f'{column}_tfidf_{j+1}' for j in range(results.shape[1])])
        train_sets[i] = pd.concat([train_sets[i], tfidf_df], axis=1)
        train_sets[i] = train_sets[i].drop(column, axis=1)
```
