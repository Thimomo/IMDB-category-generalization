```python
import pandas as pd
```

```python
X_train = pd.read_csv('Datasets/base_X_train.csv')
X_test = pd.read_csv('Datasets/base_X_test.csv')
train_sets = [X_train, X_test]
```

```python
X_train.head()
```

```python
clean_columns = ['genre', 'director', 'stars', 'writer']

for i in range(len(train_sets)):   
    for column in clean_columns:
        train_sets[i][column] = train_sets[i][column].astype(str).str.replace('[', '', regex=False).str.replace(']', '', regex=False)
        train_sets[i][column] = train_sets[i][column].apply(lambda x: [item.strip() for item in x.split(',') if item.strip()])

print(type(train_sets[0]['writer'][0]))
```
