```python
import pandas as pd
import numpy as np
import re

# Laad de dataset
file_path = 'Datasets/cleaned_X_train.csv'
df = pd.read_csv(file_path)

# Functie om een string met numerieke waarden om te zetten naar een lijst en het gemiddelde te berekenen
def process_list_column(column):
    def clean_and_eval(x):
        try:
            # Verwijder overbodige spaties en maak de string compatibel als lijst
            x = re.sub(r'\s+', ',', x.strip())  # Vervang witte ruimte door komma's
            x = f"[{x}]"  # Voeg haakjes toe zodat het een lijst wordt
            elements = eval(x)  # Converteer naar een Python-lijst
            return np.mean(elements)  # Bereken het gemiddelde
        except Exception as e:
            print(f"Fout bij het verwerken van de waarde: {x}. Foutmelding: {e}")
            return np.nan

    return column.apply(clean_and_eval)

# Pas de functie toe op de 'title' en 'IMDb description' kolommen
df['title'] = process_list_column(df['title'])
df['IMDb description'] = process_list_column(df['IMDb description'])

# Opslaan van de aangepaste DataFrame
output_file_path = 'Datasets/cleaned_X_train2.csv'
df.to_csv(output_file_path, index=False)

print(f"De gegevens zijn succesvol opgeslagen in: {output_file_path}")




```

```python
import pandas as pd
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from matplotlib import pyplot

# Laad jouw dataset (pas deze paden aan naar je eigen dataset)
X_train = pd.read_csv('Datasets/cleaned_X_train2.csv')
y_train = pd.read_csv('Datasets/y_train.csv')

# Converteer de one-hot encoded target naar klasse labels (1D array)
y_train = y_train.idxmax(axis=1)

# Verkrijg een lijst van modellen om te evalueren
def get_models():
    models = dict()
    for p in [0.0, 0.0001, 0.001, 0.01, 0.1, 1.0]:
        # Maak naam voor model
        key = '%.4f' % p
        # Zet penalty uit in sommige gevallen
        if p == 0.0:
            models[key] = LogisticRegression(multi_class='multinomial', solver='lbfgs', penalty='none')
        else:
            models[key] = LogisticRegression(multi_class='multinomial', solver='lbfgs', penalty='l2', C=p)
    return models

# Evalueer een gegeven model met cross-validation
def evaluate_model(model, X, y):
    # Definieer de evaluatieprocedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # Evalueer het model
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    
    # Train het model voor classificatierapport
    model.fit(X, y)
    y_pred = model.predict(X)
    report = classification_report(y, y_pred)
    
    return scores, report

# Verkrijg de modellen om te evalueren
models = get_models()

# Evalueer de modellen en sla de resultaten op
results, reports, names = list(), list(), list()
for name, model in models.items():
    # Evalueer het model en verzamel de scores en classificatierapport
    scores, report = evaluate_model(model, X_train, y_train)
    # Sla de resultaten op
    results.append(scores)
    reports.append(report)
    names.append(name)
    # Geef de voortgang weer
    print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
    print(f"Classification Report for {name}:\n{report}")

# Plot de modelprestaties ter vergelijking
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.title("Model Comparison: Accuracy")
pyplot.show()

```

```python
import pandas as pd
from numpy import mean, std
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report
from matplotlib import pyplot

# Laad jouw dataset
X_train = pd.read_csv('Datasets/cleaned_X_train2.csv')
y_train = pd.read_csv('Datasets/y_train.csv')

# Converteer de one-hot encoded target naar klasse labels (1D array)
y_train = y_train.idxmax(axis=1)

# Encodeer de stringlabels naar numerieke labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Schaal de gegevens
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Bereken class weights (voor ongebalanceerde datasets)
class_weights = compute_class_weight('balanced', classes=np.unique(y_train_encoded), y=y_train_encoded)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Specificeer de hyperparameters om te zoeken
param_grid = {
    'C': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0],  # Regularisatie parameter
    'solver': ['lbfgs'],  # Oplossers voor Logistic Regression (alleen lbfgs ondersteunt multinomiale)
    'penalty': ['l2'],  # L2-regularisatie
    'max_iter': [100, 200, 300]  # Aantal iteraties (indien nodig verhogen)
}

# Maak een Logistic Regression model aan
logreg = LogisticRegression(multi_class='multinomial', class_weight=class_weight_dict)

# Gebruik GridSearchCV om de beste hyperparameters te vinden
grid_search = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train_encoded)

# Toon de beste parameters en de bijbehorende score
print(f"Beste hyperparameters: {grid_search.best_params_}")
print(f"Beste score: {grid_search.best_score_}")

# Gebruik de beste parameters voor verder evalueren
best_model = grid_search.best_estimator_

# Evalueer het model met cross-validation
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(best_model, X_train_scaled, y_train_encoded, scoring='accuracy', cv=cv, n_jobs=-1)

# Toon de resultaten
print(f"Gemiddelde cross-validation accuracy: {mean(scores):.3f} (+/- {std(scores):.3f})")

# Toon het classification report
y_pred = best_model.predict(X_train_scaled)
print("\nClassification Report:")
print(classification_report(y_train_encoded, y_pred, target_names=label_encoder.classes_))

# Plot de resultaten ter vergelijking
# Vergelijk de modellen
names = ['Logistic Regression']
results = [scores]
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.title("Performance van Logistic Regression na Hyperparameter Tuning")
pyplot.show()



```
