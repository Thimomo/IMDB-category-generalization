```python
import pandas as pd
import numpy as np
import re

# Laad de dataset
file_path = 'Datasets/cleaned_X_train.csv'
df = pd.read_csv(file_path)

# Functie om een string met numerieke waarden om te zetten naar een lijst en het gemiddelde te berekenen
def process_list_column(column):
    def clean_and_eval(x):
        try:
            # Verwijder overbodige spaties en maak de string compatibel als lijst
            x = re.sub(r'\s+', ',', x.strip())  # Vervang witte ruimte door komma's
            x = f"[{x}]"  # Voeg haakjes toe zodat het een lijst wordt
            elements = eval(x)  # Converteer naar een Python-lijst
            return np.mean(elements)  # Bereken het gemiddelde
        except Exception as e:
            print(f"Fout bij het verwerken van de waarde: {x}. Foutmelding: {e}")
            return np.nan

    return column.apply(clean_and_eval)

# Pas de functie toe op de 'title' en 'IMDb description' kolommen
df['title'] = process_list_column(df['title'])
df['IMDb description'] = process_list_column(df['IMDb description'])

# Opslaan van de aangepaste DataFrame
output_file_path = 'Datasets/cleaned_X_train2.csv'
df.to_csv(output_file_path, index=False)

print(f"De gegevens zijn succesvol opgeslagen in: {output_file_path}")




```
