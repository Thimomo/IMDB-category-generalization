```python
def convert_to_numbers(data):
    converted = []
    for item in data:
        try:
            # Probeer eerst om het item als float om te zetten
            converted.append(float(item))
        except ValueError:
            # Als het geen getal is, laat het als string
            converted.append(item)
    return converted

# Converteren van de data
converted_data = convert_to_numbers(data)

# Controleer het resultaat
print(converted_data)

# Samengevoegd in een numpy array, afhankelijk van of de waarden nummers zijn
numeric_data = np.array([item if isinstance(item, (int, float)) else 0 for item in converted_data])

print(numeric_data)
```
